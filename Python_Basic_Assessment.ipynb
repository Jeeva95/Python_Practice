# 12. Write a function that accepts any number of arguments and returns their average.

```python
def average(*args) :
    total = 0
    length = len(args)
    for i in args :
        total += i
    return total / length

ans = average(1,2,3,4,5)
print(ans)
```

# 13. Write a program to count vowels, consonants, digits, and spaces in a string without using built-in count methods.

string = "Python is   so12Gr$eat  100%abc"

```python
string = "Python is so12Gr$eat 100%abc"
count_vowels = 0
count_consonants = 0
count_digits = 0
count_space = 0

for i in string :
    if i.lower() in "aeiou" :
        count_vowels += 1
        
    elif i.isalpha() :
        count_consonants += 1

    elif i.isdigit() :
        count_digits += 1

    elif i.isspace() :
        count_space += 1

print(f"Vowels : {count_vowels}, Consonants : {count_consonants}, Digits : {count_digits}, Spaces : {count_space}")
```

# 14. The ATM Retry System

Scenario: An ATM allows 3 attempts to enter a PIN.

Task:

1. Set correct_pin = 1234

2. Use a while loop to ask the user for a PIN

3. If incorrect, print "Try again"

4. If correct, print "Access Granted" and break the loop

5. If the user fails 3 times, print "Locked out" and stop the program

```python
correct_pin = 1234
count = 0

while count < 3 :
    enter_pin = int(input("Enter the pin : "))
    if enter_pin != correct_pin :
        print("Try again")
    else :
        print("Access granted")
        break
    count += 1
print("Locked out")
```

# 15. Smart Calculator Function

Scenario: A calculator app needs reusable logic.

Task:

Create a function calculate(a, b, operation)

where:

operation can be "add", "subtract", "multiply", or "divide"

Return the result instead of printing it

If division by zero is attempted, return "Invalid Operation"

```python
def calculator(a, b, operation) :
    if operation == "add" :
        print(f"Addition : {a + b}")
    elif operation == "subtraction" :
        print(f"Subtract : {a - b}")
    elif operation == "multiply" :
        print(f"Multiply : {a * b}")
    elif operation == "divide" :
        print(f"Divide : {a / b}")
a = int(input("Enter the a : "))
b = int(input("Enter the b : "))
operation = input("Enter the operation (add, subtract, multiply, divide) : ").lower()
calculator(a, b, operation)
```
