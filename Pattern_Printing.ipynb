# 10. Write a code to create the given structure using only for loop.
        *
       ***
      *****
     *******
    *********
   *********** 
   

```python
n = 5
for i in range(n) :
    for j in range(n-i-1) :
        print(" ", end = "")
    for k in range(2*i+1) :
        print("*", end = "")
    print( )
```

# 11. Write a code to create the given structure using while loop.

              |||||||||
               |||||||
                |||||
                 |||
                  |

```python
rows = 5
i = 0
while i < rows:
    
    bars = 9 - (2 * i)
    j = 0
    while j < i:
        print(" ", end="")
        j += 1
    k = 0
    while k < bars:
        print("|", end="")
        k += 1
    print()
    i += 1
```
