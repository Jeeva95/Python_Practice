# 29. Create an abstract class Ride with an abstract method fare(self, distance)
# Create two child class Bus and Car
# Bus -> 1km -> 5 rs
# Car -> 1km -> 20 rs

```python
from abc import ABC, abstractmethod
```

```python
class Ride(ABC) :
    @abstractmethod
    def fare(self, distance) :
        pass

class Bus(Ride) :
    def fare(self, distance) :
        return distance * 5

class Car(Ride) :
    def fare(self, distance) :
        return distance * 20

distance = float(input("Enter the distance in km : "))

bus = Bus()
car = Car()

print(f"Bus fare : {bus.fare(distance)} rps")
print(f"Car fare : {car.fare(distance)} rps")
```

# 30. Create a abstract class called Shape with two abstract method area(), parimeter()
# Create two child class Rectangle(length, width) and Circle(radius) and you have to complete those two methods in these both classes

```python
from abc import ABC, abstractmethod

class Shape(ABC) :
    @abstractmethod
    def area(self) :
        pass

    @abstractmethod
    def perimeter(self) :
        pass

class Rectangle(Shape) :
    def __init__(self, length, width) :
        self.length = length
        self.width = width
        
    def area(self) :
        return self.length * self.width
    def perimeter(self) :
        return 2 * (self.length + self.width)

class Circle(Shape) :
    def __init__(self, radius) :
        self.radius = radius
    def area(self) :
        return 3.14 * self.radius ** 2
    def perimeter(self) :
        return 2 * 3.14 * self.radius

rect = Rectangle(2, 3)
cicl = Circle(4)
print(f"Rectangle : Area - {rect.area()}, Perimeter - {rect.perimeter()}")
print(f"Circle : Area - {cicl.area()}, Perimeter - {cicl.perimeter()}")
```
